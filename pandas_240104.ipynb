```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# total_bill의 중간값 이하인 것만 조회
tips.loc[tips['total_bill'] <= tips['total_bill'].median(), :]

# day가 sun인 데이터만 조회
tips.loc[tips['day'] == 'Sun', :].head(1)

# tip의 평균보다 크면서 time이 Dinner인 데이터만 조회
# tips.loc[(조건1) & (조건2), :]
tips.loc[~(tips['tip']>= tips['tip'].mean()) &
    (tips['time']=='Dinner') &
    (tips['size']==3) |
    (tips['day']=='Sat'),
:]
```

```python
tips.iloc[1,1]
```

```python
tips.iloc[:5, :3]
```

```python
tips.loc[0] # 인덱스 이름이 0인 행 추출
```

```python
tips.iloc[0] # 전체 데이터프레임에서 0번째 행에 있는 값들만 추출
```
